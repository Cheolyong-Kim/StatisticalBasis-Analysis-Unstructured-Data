```python
import codecs
import pandas as pd

with codecs.open('ratings_test.txt', encoding= 'utf-8') as f:
    data = [line.split('\t') for line in f.read().splitlines()]
    m = data[0]
    data = data[1:]
data
```

```python
p_data = pd.DataFrame(data, columns = m)
p_data
```

### 단어 토큰화

```python
reviews = p_data['document'].str.replace('[^ㄱ-ㅎ ㅏ-ㅣ 가-힣]', '')
```

```python
from konlpy.tag import Okt

tk = Okt()
tk_d = []
for review in reviews:
    l = [word for word in tk.morphs(review) if len(word) > 1 and word not in stop_w]
    tk_d.append(' '.join(l))
```

```python
tk_d
```

### 데이터 전처리

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from sklearn.feature_extraction.text import TfidfVectorizer

ttk = Tokenizer(num_words = 10000)
ttk.fit_on_texts(tk_d)
ttk.word_index
```

```python
ttk.word_counts
```

```python
Y_data = p_data['label'].values
Y_data
```

```python
from tensorflow.keras.utils import to_categorical

Y_data = to_categorical(Y_data)
Y_data
```

```python
t_x_data = ttk.texts_to_matrix(tk_d, mode = 'tfidf')
t_x_data
```

```python
t_x_data.shape
```

### 데이터 분류

```python
from sklearn.model_selection import train_test_split

t_x, tt_x, t_y, tt_y = train_test_split(t_x_data, Y_data, random_state = 1)
```

### 모델 생성 및 학습

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout

m = Sequential()
m.add(Dense(64, input_shape = (10000, ), activation = 'relu'))
m.add(Dropout(0.4))
m.add(Dense(2, activation = 'sigmoid'))
m.compile(optimizer = 'rmsprop', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
hy = m.fit(t_x, t_y, epochs = 50, validation_data = (tt_x, tt_y))
```

### loss 변화 시각화

```python
import matplotlib.pyplot as plt

ec = range(1, len(hy.history['accuracy']) + 1)
plt.plot(ec, hy.history['loss'])
plt.plot(ec, hy.history['val_loss'])
plt.show()
```
