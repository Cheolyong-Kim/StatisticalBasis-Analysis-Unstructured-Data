```python
all_data = DoublespaceLineCorpus('data1.txt', iter_sent = True)  # iter_sent = 한 줄 씩 문장으로 가져옴
len(all_data)
```

```python
n = 0 
for i in all_data:
    print(i)
    n += 1
    if n == 4:
        break
```

```python
m = WordExtractor()
m.train(all_data)
```
