```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers
```

```python
X = [1, 2, 3, 4, 5, 6, 7, 8, 9]
Y = [10, 22, 30, 45, 50, 66, 75, 80, 99]
```

```python
m = Sequential()
m.add(Dense(1, input_dim = 1, activation = 'linear'))
sgd = optimizers.SGD(lr = 0.01)
m.compile(optimizer = sgd, loss = 'mse', metrics = ['mse'])
```

```python
m.fit(X, Y, epochs = 300)
```

```python
m.predict(X)
```

```python
import matplotlib.pyplot as plt

plt.plot(X, m.predict(X), 'r')
plt.scatter(X, Y)
plt.show()
```

```python
import pandas as pd

data = pd.read_csv('data3.csv')
data
```

```python
data.pop('Id')
data
```

```python
X = data.iloc[:, :-1].values
Y = data.iloc[:, -1].values
X, Y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

x_data, tt_x, y_data, tt_y = train_test_split(X, Y, random_state = 42)
t_x, v_x, t_y, v_y = train_test_split(x_data, y_data, random_state = 42, test_size = 0.2)
```

```python
lo_g = LogisticRegression(multi_class= 'multinomial', solver= 'sag')
lo_g.fit(t_x, t_y)
```

```python
lo_g.score(v_x, v_y)
```

```python
lo_g.score(t_x, t_y), lo_g.score(tt_x, tt_y)
```

```python
from sklearn.metrics import classification_report

y_t = tt_y.copy()
y_p = lo_g.predict(tt_x)

print(classification_report(y_t, y_p))
```
