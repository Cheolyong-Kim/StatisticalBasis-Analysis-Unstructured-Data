```python
def cos_i(A, B):
    return np.dot(A, B) / (np.linalg.norm(A) * np.linalg.norm(B))
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

data = pd.read_csv('data1.csv', low_memory = False)
data = data.head(20000)
data['overview'] = data['overview'].fillna('')
data[:50]
```

```python
tfidf = TfidfVectorizer(stop_words = 'english')
tfidf_t = tfidf.fit_transform(data['overview'])
tfidf_t
```

```python
cos = cosine_similarity(tfidf_t, tfidf_t)  # Î¶¨Î∑∞ 1Í∞ú Îãπ Î™®Îì† Î¶¨Î∑∞ ÎπÑÍµêÌïòÏó¨ ÏΩîÏÇ¨Ïù∏ Ïú†ÏÇ¨ÎèÑ ÌôïÏù∏
cos.shape
```

```python
t_idx = dict(zip(data['title'], data.index))
t_idx
```

```python
def ck_s_t(t, cosine_sim = cos):
    idx = t_idx[t]
    c_sc = list(enumerate(cosine_sim[idx]))
    c_sc = sorted(c_sc, key = lambda x: x[1], reverse = True)
    m_i = c_sc[1:6]  # ÏΩîÏÇ¨Ïù∏ Ïú†ÏÇ¨ÎèÑ Í≤ÄÏÇ¨Î•º Ìï† Îïå ÏûêÍ∏∞ ÏûêÏã†Ïùò Î¶¨Î∑∞ÎèÑ Í≤ÄÏÇ¨ÌñàÍ∏∞ ÎïåÎ¨∏Ïóê Ïú†ÏÇ¨ÎèÑÍ∞Ä Í∞ÄÏû• ÎÜíÏùÄ Î¶¨Î∑∞ ÌïòÎÇòÎäî Ï†úÏô∏ÌïòÍ≥† Ïú†ÏÇ¨Ìïú Î¶¨Î∑∞ 5Í∞úÎßå
    m_idx = [i[0] for i in m_i]  # Î¶¨Î∑∞ÏôÄ Ïú†ÏÇ¨Ìïú ÏòÅÌôî Ï†úÎ™©
    return data['title'].iloc[m_idx]
```
