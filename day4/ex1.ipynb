```python
def cos_i(A, B):
    return np.dot(A, B) / (np.linalg.norm(A) * np.linalg.norm(B))
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

data = pd.read_csv('data1.csv', low_memory = False)
data = data.head(20000)
data['overview'] = data['overview'].fillna('')
data[:50]
```

```python
tfidf = TfidfVectorizer(stop_words = 'english')
tfidf_t = tfidf.fit_transform(data['overview'])
tfidf_t
```

```python
cos = cosine_similarity(tfidf_t, tfidf_t)  # 리뷰 1개 당 모든 리뷰 비교하여 코사인 유사도 확인
cos.shape
```

```python
t_idx = dict(zip(data['title'], data.index))
t_idx
```

```python
def ck_s_t(t, cosine_sim = cos):
    idx = t_idx[t]
    c_sc = list(enumerate(cosine_sim[idx]))
    c_sc = sorted(c_sc, key = lambda x: x[1], reverse = True)
    m_i = c_sc[1:6]  # 코사인 유사도 검사를 할 때 자기 자신의 리뷰도 검사했기 때문에 유사도가 가장 높은 리뷰 하나는 제외하고 유사한 리뷰 5개만
    m_idx = [i[0] for i in m_i]  # 리뷰와 유사한 영화 제목
    return data['title'].iloc[m_idx]
```
