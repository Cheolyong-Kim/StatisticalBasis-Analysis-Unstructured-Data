### 1. 데이터 로드를 통한 데이터 프레임 완성

```python
import pandas as pd

data = pd.read_csv('data2.csv')
data
```

### 2. 형태소 분석기를 이용하여 단어 토큰화 작업 후 TF-IDF를 계산

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from konlpy.tag import Okt

# 단어 토큰화
tk = Okt()
tk_l = []
for sentence in data['content']:
    l = [word for word in tk.morphs(sentence) if len(word) > 1]  # 길이가 1이하인 단어 제외
    tk_l.append(' '.join(l))
tk_l
```

```python
# TF-IDF 계산
tfidf_v = TfidfVectorizer()
tfidf_m = tfidf_v.fit_transform(tk_l)
```
