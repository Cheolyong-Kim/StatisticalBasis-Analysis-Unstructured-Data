```python
from sklearn.datasets import load_boston

data = load_boston()
print(data.data)
print(data.feature_names)
print(data.target)
```

```python
import pandas as pd

df = pd.DataFrame(data.data, columns = data.feature_names)
df['target'] = data.target
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for i in range(3):
    k = list(data.feature_names[4 * i:4 + (4 * i)]) + ['target']
    sns.pairplot(vars = k, data = df)
    plt.show()
```

```python
for i in data.feature_names:
    plt.scatter(df['target'], df[i])
    plt.xlabel('target')
    plt.ylabel(i)
    plt.show()
```
