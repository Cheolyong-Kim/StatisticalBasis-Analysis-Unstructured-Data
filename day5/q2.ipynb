```python
X_data = data[['1', '2', '7']].values
Y_data = data['Y'].values
```

```python
from tensorflow.keras.utils import to_categorical

Y_data=to_categorical(Y_data)
Y_data
```

### ë°ì´í„° ì „ì²˜ë¦¬

```python
from sklearn.model_selection import train_test_split

x_data, tt_x, y_data, tt_y = train_test_split(X_data, Y_data, random_state = 1)
t_x, v_x, t_y, v_y = train_test_split(x_data, y_data, random_state = 1, test_size = 0.2)
```

### ëª¨ë¸ ìƒì„± ë° í•™ìŠµ

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

m = Sequential()
m.add(Dense(3, input_dim = 3, activation = 'softmax'))
m.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
hy = m.fit(t_x, t_y, epochs = 300, batch_size = 1, validation_data = (v_x, v_y))
```

### lossê°’ ì‹œê°í™”

```python
import matplotlib.pyplot as plt

ec = range(1, len(hy.history['accuracy']) + 1)
plt.plot(ec, hy.history['loss'], label = 'loss')
plt.plot(ec, hy.history['val_loss'], label = 'val_loss')
plt.legend()
plt.show()
```

### accuracyê°’ ì‹œê°í™”

```python
ec = range(1, len(hy.history['accuracy']) + 1)
plt.plot(ec, hy.history['accuracy'], label = 'accuracy')
plt.plot(ec, hy.history['val_accuracy'], label = 'val_accuracy')
plt.legend()
plt.show()
```

```python
m.evaluate(tt_x, tt_y)
```
